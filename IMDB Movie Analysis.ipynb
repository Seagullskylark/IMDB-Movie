* classify Movies Based on Rating [Excellent,Good or Average]

```python
data.columns
```

```python
def rating(rating):
    if rating >=8.0:
        return 'Excellent'
    elif rating >= 7.0:
        return 'Good'
    else :
        return 'Average'
```

```python
data['Rating_Category']=data['Rating'].apply(rating)
```

* Find Unique values from Genre

```python
data.columns
```

```python
data['Genre']
```

```python
list1=[]

for i in data['Genre']:
    list1.append(i.split(','))
```

```python
list1
```

```python
l1=[]
for i in list1:
    for j in i:
        l1.append(j)
```

```python
l1
```

```python
uni=[]
for i in l1:
    if i not in uni:
        uni.append(i)
```
